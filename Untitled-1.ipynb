```python
import pandas as pd

dfData = pd.read_csv('train.csv')
dfData.head(5)
```

```python
# Отфильтруем данные
df = dfData[(dfData['Store'] == 1) & (dfData['Dept'] == 1)]

# Преобразуем дату с формат datetime
df['Date'] = pd.to_datetime(df['Date'])

# Удалим лишние столбцы
df = df.drop('Store', axis=1)
df = df.drop('Dept', axis=1)
df = df.drop('IsHoliday', axis=1)

# Переименуем столбцы
df = df.rename(columns={'Date': 'timestamp', 'Weekly_Sales': 'Value'})

# Повесим индексы на столбец timestamp
#df.set_index('timestamp', inplace=True)

# Агрегация по месяцам
df = df.groupby(df['timestamp'].dt.strftime('%Y-%B'))['Value'].sum()

print(df.head(5))
```

```python
# Проверка на стационарность ряда (расширенный тест Дики-Фуллера ADF)
from statsmodels.tsa.stattools import adfuller
import matplotlib.pyplot as plt

result = adfuller(df)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])

# Визуализация скользящих статистик
rolling_mean = pd.Series(df).rolling(window=12).mean()
rolling_std = pd.Series(df).rolling(window=12).std()
    
plt.figure(figsize=(12,6))
plt.plot(df, label='Исходный ряд')
plt.plot(rolling_mean, label='Скользящее среднее')
plt.plot(rolling_std, label='Скользящее стандартное отклонение')
plt.legend()
plt.title('Анализ стационарности временного ряда')
plt.show()
```

```python
# ARMA - сумма моделей скользящего среднего и авторегрегсии. Используется для стационарных временных рядов (рядов, у которых среднее значение, дисперсия и прочие хар-ки не изменяются во времени)
```

```python
# Базовая структура ARIMA. Если ряд не стационарен - применяем дифференцирование. Если дифференционный ряд стационарен - то d = 1
import numpy as np

class datArima:
    def __init__(self, p, d, q):
        self.p = p
        self.d = d
        self.q = q
        self.ar_params = np.zeros(p)
        self.ma_params = np.zeros(q)

    def difference(self,series, d=1):
        """Выполняет d-кратное дифференцирование ряда"""
        diff = series.copy()
        for _ in range(d):
            diff = np.diff(diff)
        return diff

    def inverse_difference(self, diff, original, d=1):
        """Восстанавливает исходный ряд после дифференцирования"""
        restored = diff.copy()
        for _ in range(d):
            restored = np.cumsum(restored) + original[0]
        return restored
```

```python
# Автоматический подбор параметров модели ARIMA
import itertools
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error

def grid_search_arima_params(data, p_range, d_range, q_range):
    best_score = float('inf')
    best_params = None
    
    for p, d, q in itertools.product(p_range, d_range, q_range):
        try:
            model = ARIMA(data, order=(p, d, q))
            results = model.fit()
            aic = results.aic
            
            if aic < best_score:
                best_score = aic
                best_params = (p, d, q)
                
        except:
            continue
            
    return best_params, best_score

# Пример использования
p_range = range(0, 4)
d_range = range(0, 3)
q_range = range(0, 4)

best_params, best_score = grid_search_arima_params(
    df,
    p_range,
    d_range,
    q_range
)

print(f"Лучшие параметры (p,d,q): {best_params}")
print(f"AIC score: {best_score}")
```

```python
# Создаем модель и подставляем лучшие параметры
model = ARIMA(df, order=(best_params[0],best_params[1],best_params[2]))
model_fit = model.fit()
print(model_fit.summary())
```

```python
# Диагностика модели с помощью анализа автокорреляционной и частичной автокорреляционной функций остатков. Если значимых пиков нет, то всё ок
import statsmodels.api as sm
residuals = model_fit.resid
fig, ax = plt.subplots(1,2, figsize=(15,5))
sm.graphics.tsa.plot_acf(residuals, lags=30, ax=ax[0])
sm.graphics.tsa.plot_pacf(residuals, lags=30, ax=ax[1])
plt.show()
```

```python
# Визуализация результатов модели на обучающей выборке
forecast = model_fit.forecast(steps=len(df))

plt.figure(figsize=(12, 6))
plt.plot(df, label='Обучающая выборка')
plt.plot(df.index, forecast, label='Прогноз', color='red')
plt.title('ARIMA прогноз')
plt.legend()
plt.show()
```
